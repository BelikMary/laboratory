№ 5.2.6

Дан список из чисел.

Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).


```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_01.
#
# Выполнил: Белик М.К.
# Группа: АДЭУ-211

import  math

def gcd(first, second):
    return math.gcd(first, second)


def lcm(first, second):
    lcm.multiple = lcm.multiple + second
    if ((lcm.multiple % first == 0) and (lcm.multiple % second == 0)):
        return lcm.multiple;
    else:
        lcm(first, second)
    return lcm.multiple
lcm.multiple = 0


def gcd_nums(nums):
    gcd_nums_ = nums[0]
    for i in range(1, len(nums)):
        gcd_nums_ = gcd(gcd_nums_, nums[i])
    return gcd_nums_


def lcm_nums(nums):
    lcm_nums_ = nums[0]
    for i in range(1, len(nums)):
        lcm_nums_ = lcm(lcm_nums_, nums[i])
    return lcm_nums_


nums = [int(i) for i in input('Введите числа через пробел: ').split(sep=' ')]
print("НОД = ", gcd_nums(nums))
print("НОK = ",lcm_nums(nums))


# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```
