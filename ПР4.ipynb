```python
class Style:

    def __init__(self, words, language_rus):
        self.words=words
        self.language_rus=language_rus
        
    def translate(self):
        if self.language_rus == True:
           return ('Произведение не нуждается в переводе,')
        else:
           return ('Произведение нуждается в переводе,')

    def kassa(self):
        if self.language_rus == True:
          return self.words*0.02
        else:
          return self.words*0.04

class Proza(Style):

    def __init__(self, words, language_rus, words_on_page):
        Style.__init__(self, words, language_rus)
        self.words_on_page = words_on_page

    def pages(self):
        return round(self.words/self.words_on_page,0)

class Stih(Style):

    def __init__(self, words, language_rus, words_on_line):
        Style.__init__(self, words, language_rus)
        self.words_on_line = words_on_line

    def pages(self):
        return round((self.words/self.words_on_line)/20,0)

class Rifma(Stih):

    def __init__(self, words, language_rus, words_on_line, if_rifma_is_good):
      Stih.__init__(self,words, language_rus, words_on_line)
      self.if_rifma_is_good = if_rifma_is_good

    def nazenka(self):
      if self.if_rifma_is_good == True:
        return Style.kassa(self)*10
      else:
        return Style.kassa(self)*5
```

```python
Remark=Proza(100000, False, 300)
print(Remark.translate(),'поэтому цена книги составила ', Remark.kassa(), ' (количество страниц ', Remark.pages(), ')')

Lermontov=Stih(420, True, 4)
print(Lermontov.translate(),'поэтому цена книги составила ', Lermontov.kassa(), ' (количество страниц ', Lermontov.pages(), ')')

Mayakovsky = Rifma(420, True, 3,False)
print(Mayakovsky.translate(),'поэтому цена книги составила ', Mayakovsky.nazenka(), ' (количество страниц ', Mayakovsky.pages(), ')')
```
